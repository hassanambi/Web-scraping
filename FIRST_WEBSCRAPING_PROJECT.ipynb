```python
#importing the library

from bs4 import BeautifulSoup
import requests
```

```python
#collecting the webpage

url = 'https://en.wikipedia.org/wiki/List_of_largest_energy_companies'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
 # trying to locate the table

 soup.find_all('table')
```

```python
#using class to locate the exact table

table = soup.find('table', class_='sortable wikitable')
print(table)
```

```python
#extracting the columns(headers)

table_cols = table.find_all('th')
for col in table_cols:
  print(col.text.strip())
```

```python
#using list comprehension

table_cols = [col.text.strip() for col in table.find_all('th')]
print('Table Headers:')
print(table_cols)
```

```python
#extracting the row data
print("Extracted data")

table_rows = table.find_all('tr')
rows=[]

for row in table_rows:
  cells = row.find_all('td')
  data = [cell.text.strip() for cell in cells]
  if data:
    rows.append(data)
    print(data)
```

```python
#creating a dataframe

import pandas as pd

df = pd.DataFrame(rows, columns= table_cols)
```

```python
df.to_csv('Top_10_energy_company_in-world.csv', index=False)
```
